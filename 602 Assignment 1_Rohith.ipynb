Generate a normally distributed data, with mean=3 and standard deviation=2, for 10,000 observations:

```python
import numpy as np
mu, sigma = 3, 2 
real = np.random.normal(mu, sigma, 10000)
real
```

Plot the normal distribution overlaid on top of a histogram in matplotlib

```python
import matplotlib.pyplot as x
import pandas as pd
from matplotlib import pyplot as plt
sp = pd.Series(real)
sp.plot.kde()
plt.hist(real,density = True)
```

Create a sub-sample of this population with mean of  1.5, a sigma of 2.0, and 200 observations. 

```python
import numpy as np
mean_sample, sigma_sample = 1.5, 2
sample = np.random.normal(mean_sample, sigma_sample, 200)
```

Compare both distributions on a chart

```python
count, bins, ignored = plt.hist(sample, 30, alpha=0.1, density=True)
pd.Series(sample).plot.kde(color = 'r') #Using red for sample distribution
count, bins, ignored = plt.hist(real, 30, alpha=0.1, density=True)
sp.plot.kde(color = 'b') #Using blue for real distribution
```

Use ‘from scipy.stats import norm’ to create 95% confidence intervals

```python
from scipy.stats import norm
intp_95 = norm.interval(0.95, loc=3, scale=2)
ints_95 = norm.interval(0.95, loc=1.5, scale=2)
count, bins, ignored = plt.hist(sample, 30, alpha=0.1, density=True)
pd.Series(sample).plot.kde(color = 'r')
count, bins, ignored = plt.hist(real, 30, alpha=0.1, density=True)
sp.plot.kde(color = 'b')
plt.axvline(ints_95[1],color='g')
plt.axvline(ints_95[0],color='g')
```

Compute the z-score, which is equal to (sample mean – mu)/sigma

```python
sample_mean = sample.mean()
real_mean = real.mean()
from scipy.stats import norm
real_v = real.var()
z = (sample_mean - real_mean)/real_v
print("z-score:",z)
p = norm.sf(abs(z))
print("p-value:",p)
```

Since p-value is greater than 0.05, it is weak evidence against the null hypothesis. Hence, we cannot reject the null hypothesis.
